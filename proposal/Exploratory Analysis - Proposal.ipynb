```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas
import json
```

```python
df = pandas.read_csv('tmdb_5000_movies.csv')
```

```python
df_movies = df[['id','budget', 'title', 'release_date', 'revenue', 'vote_average', 'vote_count']]
```

```python
df_movies
```

```python
credits = pandas.read_csv('tmdb_5000_credits.csv')
```

```python
df_credits = credits[['movie_id','cast']]
```

```python
df_credits
```

```python
df = df_movies.join(df_credits, lsuffix='id', rsuffix='movie_id')
```

```python
df
```

```python
x=df['budget']
y=df['revenue']

ax1 = plt.scatter(x=x, y=y, c='DarkBlue')

plt.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))(np.unique(x)))

plt.xlabel("Budget (Hundreds of Millions)")
plt.title("Movie Budget vs Revenue")
plt.ylabel("Revenue(Billions)")
plt.show()
```

```python
x=df['vote_average']
y=df['revenue']

ax2 = plt.scatter(x=x, y=y, c='Black')
plt.xlabel("Average Rating")
plt.title("Movie Rating vs Revenue")
plt.ylabel("Revenue(Billions)")
plt.show()
```

```python
x=df['vote_count']
y=df['revenue']

ax3 = plt.scatter(x=x, y=y, c='Red')

plt.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1,))(np.unique(x)))
plt.xlabel("Number of Votes")
plt.title("Movie Vote Count vs Revenue")
plt.ylabel("Revenue(Billions)")
plt.show()
```
