```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('tmdb_5000_movies.csv')
```

```python
df_movies = df[['id','budget', 'title', 'release_date', 'revenue', 'vote_average', 'vote_count']]
```

```python
df_movies
```

```python
credits = pandas.read_csv('tmdb_5000_credits.csv')
```

```python
df_credits = credits[['movie_id','cast']]
```

```python
df_credits
```

```python
df = df_movies.join(df_credits, lsuffix='id', rsuffix='movie_id')
```

```python
df
```

```python
x=df['budget']
y=df['revenue']

ax1 = plt.scatter(xlabel = 'Budget', ylabel = 'Revenue', title = 'Title', x=x, y=y, c='DarkBlue')


plt.title = "title"


plt.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))(np.unique(x)))

```
